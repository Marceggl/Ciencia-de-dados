# Visualização de Dados em Python (matplotlib)

## Importando Biblioteca Gráfica - matplotlib

```python
from matplotlib import pyplot as plt

# Consulte a documentação em https://matplotlib.org/stable/index.html
```

### Exemplo Simples

```python
x = [1, 2, 3, 4, 5, 6, 7, 8] ### padrão crescnte
y = [2, 3, 1, 3, 1, 4, 2, 3] ### Sem padrão

plt.scatter(x, y)
plt.grid(True)
```

### Gráfico de linha simples

```python
anos = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
pib =  [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# cria um gráfico em linhas, anos no eixo x, PIB no eixo y
plt.plot(anos, pib);
```

```python
# Manipulando parâmetros do gráfico
plt.plot(anos, pib, color='green', marker='o', linestyle='solid')

# Adiciona Título
plt.title("PIB Nominal")

# Adiciona legenda ao eixo y
plt.ylabel("Bilhões of R$")
```

### Gráfico de Barras

Um gráfico de barras é recomendado quanto for necessário mostrar como uma grandeza quantitativa varia ao longo de itens discretos

```python
filmes = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# gera um gráfico de barras com as coordenada horizontal em [0, 1, 2, 3, 4], altura [num_oscars]
plt.bar(range(len(filmes)), num_oscars);

plt.title("Meus Filmes Favoritos")  # Adiciona Título
plt.ylabel("# de Oscars")   # Legenda eixo y

# legenda eixo x com nome dos filmes ao centro
plt.xticks(range(len(filmes)), filmes);
```

### Gráfico de Setor (Pizza)

Um gráfico de setores é recomendado quanto for necessário comparar proporções e distribuições entre os dados.

```python
plt.pie(num_oscars, labels=filmes, autopct='%1.1f%%');
```

### Histograma a partir do gráfico de barras

Um gráfico de barras pode ser uma escolha para construir histogramas para valores numéricos agrupados, afim de explorar visualmente como estes valores estão agrupados.

```python
from collections import Counter
notas = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# agrega notas por dezena, mas coloca o 100 dentro dos 90s
histograma = Counter(min(nota // 10 * 10, 90) for nota in notas)

plt.bar([x + 5 for x in histograma.keys()],  # desloca as barras em 5 para a direita
        histograma.values(),                 # Atribui a cada barra a altura correta
        10,                                  # Atribui a cada barra a largura de 8
        edgecolor=(0, 0, 0))                 # bordas pretas em cada barra

plt.axis([-5, 105, 0, 5])                  # x-axis de -5 a 105,
                                           # y-axis de 0 a 5

plt.xticks([10 * i for i in range(11)])    # rotulos do eixo x em 0, 10, ..., 100
plt.xlabel("Dezena")
plt.ylabel("# de estudantes ")
plt.title("Distribuição das notas da Prova 1");
```

### Histograma a partir da função 'hist' **(mais fácil)**

```python
notas = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# Gera exatamente o mesmo gráfico que o item anterior a partir da função hist
plt.hist(notas, edgecolor=(0, 0, 0));

plt.xlabel("Dezena")
plt.ylabel("# de estudantes ")
plt.title("Distribuição das notas da Prova 1");
```

### Gráfico de Dispersão (Scatterplot)

Um gráfico de dispersão pode ser uma escolha ao visualizar a relação entre pares de dados relacionados.

```python
amigos = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutos = [175, 170, 205, 120, 220, 130, 105, 145, 190]


plt.scatter(amigos, minutos)

plt.title("Relação de tempo no pátio e número de amigos")
plt.xlabel("número de amigos")
plt.ylabel("minutos diários gastos no pátio");
```

```python
rotulos =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(amigos, minutos)

# rótulo para cada ponto
for label, friend_count, minute_count in zip(rotulos, amigos, minutos):
    plt.annotate(label,
        xy=(friend_count, minute_count), # Põe rotulo no ponto
        xytext=(5, -5),                  # coordenada ligeiramente diferente do ponto
        textcoords='offset points')

plt.title("Minutos Diários vs. Número de Amigos")
plt.xlabel("número de amigos")
plt.ylabel("minutos diários gastos no pátio");
```

## Gerando um sinal e plotando gráficos

```python
import numpy as np
```

### Gerando um vetor de 50 posições, igualmente "espaçadas", com valores de 0 a 6$\pi$

```python
x = np.linspace(0, 6 * np.pi, 50)
```

### Gerando um sinal senoidal a partir das posições do vetor anteriormente criado

```python
y = np.sin(x)
```

### Plotando o sinal

```python
plt.plot(x, y, label='sin(t)')
plt.legend()
plt.title('Sinal Senoidal')
plt.xlabel('Tempo')
plt.ylabel('Valor')
```

### Plotando o gráfico de dispersão ('Scatter')

```python
plt.scatter(x, y, label='cos(t)')
plt.legend()
plt.title('Sinal Senoidal')
plt.xlabel('Tempo')
plt.ylabel('Valor')
```

### Plotando múltiplos gráficos em uma única figura

```python
plt.plot(x, y, label='cos(t)')
plt.plot(x, 2 * y, label='2 * cos(t)')
plt.legend()
plt.title('Sinais Senoidais')
plt.xlabel('Tempo')
plt.ylabel('Valor')
```

```python
x = np.linspace(0, 2, 100)  # Dados

fig, axs = plt.subplots(2, 2)  # grade 2x2
axs[0,0].plot(x,x, label='linear')
axs[0,1].plot(x,x**2, label='quadratico')
axs[1,0].plot(x,x**3, label='cubico')
axs[1,1].plot(x,np.random.randn(100,), label='aleatório')
# axs[1,1].plot(x,np.sqrt(x), label='raiz quadrada')
```

### C) Gráfico 3

```python
legenda = ['Sci-fi', 'Comedy', 'Action', 'Romance']
valores = [25, 15, 35, 45]

# Coloca apenas os valores inteiros no gráfico
def valores_def(pct, allvals):
  absolute = int(np.round(pct/100.*np.sum(allvals)))
  return f"{absolute:d}"

fig, ax = plt.subplots(figsize=(4,3.2))

ax.pie(valores,
       labels=None,
       colors=['#FF8629', '#FCFF34', '#EA3856', '#3AAA90'],
       autopct=lambda pct:valores_def(pct, valores),
       startangle=90,
       counterclock=False,
       wedgeprops = {"edgecolor" : "black",
                      'linewidth': 1,
                      'antialiased': True},
       )

ax.legend(legenda,
          loc='upper right',
          bbox_to_anchor=(0.8, 0.36,0.5,0.5),
          frameon=False
          )

plt.show()
```
