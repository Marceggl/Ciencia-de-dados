# Classificação com Decision Tree - ScikitLearn

## Parte 1 - Exemplo de Classificação

### Gerando conjunto de dados para classificação

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=100, centers=2, n_features=2, random_state=0)

X.shape # features
```

```python
print(y) # rotulos
```

```python
from matplotlib import pyplot as plt

# ESCREVA AQUI O CÓDIGO QUE GERA UM GRÁFICO DE DISPERSÃO (SCATTER) QUE RELACIONA OS ATRIBUTOS (X)
plt.scatter(X[:,0],X[:,1], c=y)

plt.title("Espaço de Atributos")
plt.ylabel("Atributo 1 (X1)")
plt.xlabel("Atributo 2 (X0)")
```

### Divisão para treino e teste

70% para treino e 30% para testes

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Plot apenas do conjunto de TREINO

```python
plt.scatter(X_train[:,0],X_train[:,1], c=y_train)
plt.scatter(X_test[:,0],X_test[:,1], c=y_test, cmap='winter', alpha=0.5)

plt.title("Conjunto de Treino")
plt.ylabel("Atributo 1")
plt.xlabel("Atributo 2")
```

### Classificação - Algoritmo Decision Tree (Árvore de Decisão)

#### Importando algoritmo de Árvore de Decisão do scikitlearn




```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier();
```

### Treinamento

```python
dt.fit(X_train,y_train); # Treinando o Algoritmo Decision Tree
```

### Teste

```python
dt_y_pred = dt.predict(X_test) # Classificando dados "novos" - Teste
print(dt_y_pred)
print(y_test)
```

Acurácia

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,dt_y_pred))
# print(y_test-dt_y_pred)
# print(1 - np.sum(np.abs(y_test-KNN_y_pred))/np.size(y_test))
```

### Fronteira de decisão - KNN

```python
# from sklearn.inspection import DecisionBoundaryDisplay
# DecisionBoundaryDisplay.from_estimator(KNN, X, cmap=cm, alpha=0.8, ax=ax, eps=0.5) # sklearn >v1.1.2

import numpy as np
h = 0.02  # step size in the mesh

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = dt.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)
```

```python
# Plot the training points
cm = plt.cm.RdBu
plt.contourf(xx, yy, Z, alpha=0.6, cmap=cm)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm, edgecolors="k")
plt.title("Dados de Treino")
```
